```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

output_notebook()
```

```python
from bokeh.layouts import gridplot

plot_options = dict(width=300, plot_height=250, tools='pan,wheel_zoom')

# create a new plot
s1 = figure(**plot_options, title="United States")
s1.xaxis.axis_label = "Year"
s1.yaxis.axis_label = "Aged 15+ employment rate (%)"
# s1.circle(df.columns.values, df.loc['United States'], size=6, color="navy")
s1.line(df.columns.values, df.loc['United States'], line_width=2, color="navy")

# create a new plot and share both ranges
s2 = figure(x_range=s1.x_range, y_range=s1.y_range, **plot_options, title="China")
s2.xaxis.axis_label = "Year"
# s2.yaxis.axis_label = "Aged 15+ employment rate (%)"
# s2.triangle(df.columns.values, df.loc['China'], size=6, color="firebrick")
s2.line(df.columns.values, df.loc['China'], line_width=2,color="firebrick")

# create a new plot and share only one range
s3 = figure(x_range=s1.x_range, y_range=s1.y_range, **plot_options, title="India")
s3.xaxis.axis_label = "Year"
# s3.yaxis.axis_label = "Aged 15+ employment rate (%)"
# s3.square(df.columns.values, df.loc['India'], size=6, color="olive")
s3.line(df.columns.values, df.loc['India'], line_width=2, color="olive")

p = gridplot([[s1, s2, s3]])

            
# show the results

show(p)
```

```python
from bokeh.palettes import viridis
from bokeh.layouts import column
from bokeh.models import CustomJS, ColumnDataSource, Slider

slider = Slider(start=1991, end=2007, value=1, step=1, title="Year")

countries=df.index.values.tolist()
p = figure(x_range=countries[0:20], width=950, plot_height=400, title="Aged 15+ employment rate (%)")
p.vbar(x=countries[0:20], top=df[1991][0:20], width=0.9, color=viridis(20))
p.xaxis.major_label_orientation = 120


show(column(slider, p))
# show(p)
```
